# Predictions and Analysis

## Import Libraries

```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

## Predict Housing Prices

Import model and variables.

```python
df = pd.read_csv('../datasets/clean_data.csv')
```

```python
with open('../Pickle/model.pickle', 'rb') as handle:
    lasso = pickle.load(handle)

with open('../Pickle/X_test.pickle', 'rb') as handle:
    X_test = pickle.load(handle)

with open('../Pickle/y_test.pickle', 'rb') as handle:
    y_test = pickle.load(handle)
```

Predicting housing prices with linear model.

```python
pred = lasso.predict(X_test)
```

## Analysing

Plot predicted value against actual value.

```python
plt.figure(figsize=(10,6),dpi=250)
sns.scatterplot(x=pred, y=y_test);
```

Calculate $R^2$ score.

```python
r2_score(y_test, pred)
```

Plot Coefficients for Neighborhoods Bloomington as refernce.

```python
plt.figure(figsize=(18,10),dpi=250)
sns.barplot(x=lasso.coef_[4:31], y=df.columns[4:31]);
```

## Summary

#### Predictions

The linear regression model is capable of predicting housing prices realiably, as the $R^2$ score is 0.89. Which means 89% of the variability in the target value is explained by the model. Also, from the plot we can see that predictions are less varied at the lower range and variability increase with the increase with price. 

#### Coefficients

By studying the coefficients of the model, we can inference the most expansive neighbourhoods in Ames, IA:

* Stone Brook
* Northridge Heights
* Green Hills
